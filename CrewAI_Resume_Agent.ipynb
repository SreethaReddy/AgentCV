## Extracting Resume and Job Description

````python
from PyPDF2 import PdfReader

def extract_text_from_pdf_pypdf2(pdf_path):
    try:
        reader = PdfReader(pdf_path)
        text = ""
        for page in reader.pages:
            extracted_page_text = page.extract_text()
            if extracted_page_text:
                text += extracted_page_text + "\n"
        return text
    except FileNotFoundError:
        return f"Error: The file at path '{pdf_path}' was not found."
    except Exception as e:
        return f"An error occurred: {e}"

resume = extract_text_from_pdf_pypdf2(resume_path)
````

````python
def extract_text_from_txt(txt_path):
    try:
        with open(txt_path, "r", encoding="utf-8") as f:
            text = f.read()
        return text
    except FileNotFoundError:
        return f"Error: The file at path '{txt_path}' was not found."
    except Exception as e:
        return f"An error occurred: {e}"

jd = extract_text_from_txt(jd_path)
````

## First Iteration

````python
import os
from dotenv import load_dotenv
from crewai import Agent, Task, Crew, LLM
from langchain_google_genai import ChatGoogleGenerativeAI

# Load API key
load_dotenv()
api_key = "API"

# Gemini LLM setup
llm = LLM(
    model='gemini/gemini-1.5-flash',
    api_key=api_key,
    temperature=0.0  # Lower temperature for more consistent results.
)


# Define Agents
resume_analyzer = Agent(
    role="Resume Analyzer",
    goal="Extract key strengths and weak points from the resume",
    backstory="You're an expert in resume evaluation and personal branding.",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

job_matcher = Agent(
    role="Job Matcher",
    goal="Match resume against job description and identify gaps",
    backstory="You're an expert recruiter who checks how well a resume aligns with a job.",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

# Define Tasks
analysis_task = Task(
    description=f"Analyze the following resume and highlight its key strengths and areas of improvement:\n\n{resume}",
    agent=resume_analyzer,
    expected_output="A detailed analysis of the resume's strengths and weaknesses"
)

match_task = Task(
    description=f"Compare this resume:\n{resume}\n\nwith this job description:\n{jd}\n\nHighlight missing skills, mismatches, and improvements.",
    agent=job_matcher,
    expected_output="A comparison report highlighting gaps between the resume and job requirements"
)

# Build Crew
crew = Crew(
    agents=[resume_analyzer, job_matcher],
    tasks=[analysis_task, match_task],
    verbose=True
)

# Run
result = crew.kickoff()
````
